```python
# A1. Write your own functions to evaluate the following for vectors, A & B where A & B are vectors in N-dimensional space
# Dot product between A & B.
# Length of Vectors with Euclidean Norm 
# Compare the results obtained using numpy.dot() & numpy.linalg.norm()
import numpy as np

def dot_product(vector_a,vector_b):
    dot_vector=np.dot(vector_a,vector_b)
    return dot_vector
# This function computes the dot product of two N-dimensional vectors

def length_euclidean_norm(vector_a):
    length_euclidean_norm_vector=np.linalg.norm(vector_a)
    return length_euclidean_norm_vector
# This function computes the Euclidean norm (length) of a vector

def main():
    vector_a=np.array([1,2,3])
    vector_b=np.array([4,5,6])

    print(dot_product(vector_a,vector_b))
    print(length_euclidean_norm(vector_a))
    print(length_euclidean_norm(vector_b))

if __name__=="__main__":
    main()
```

```python
# A2. Evaluate the intraclass spread and interclass distances between the classes in your dataset. If your data deals with multiple classes, you can take any two classes. Steps below (refer below diagram for understanding):
# Write your function for calculation of mean, variance and standard deviation for a set of data. Use this functions to develop a function for calculation of the values for a dataset represented as a matrix with columns as features.
# Calculate the mean for each class (also called as class centroid)
# (Suggestion: You may use numpy.mean() function for finding the average vector for all vectors in a given class. Please define the axis property appropriately to use this function. EX: feat_vecs.mean(axis=0))
# Calculate spread (standard deviation) for each class (Suggestion: You may use numpy.std() function for finding the standard deviation vector for all vectors in a given class. Please define the axis property appropriately to use this function.)
# Calculate the distance between mean vectors between classes. (Suggestion: numpy.linalg.norm(centroid1 â€“ centroid2) gives the Euclidean distance between two centroids.)
import numpy as np

def mean_vector(data):
    return np.sum(data,axis=0)/data.shape[0]
# This function computes the mean vector of a dataset along feature dimensions

def variance_vector(data):
    mu=mean_vector(data)
    return np.sum((data-mu)**2,axis=0)/data.shape[0]
# This function computes the variance vector of a dataset using the mean vector

def std_vector(data):
    return np.sqrt(variance_vector(data))
# This function computes the standard deviation vector of a dataset

def class_centroid(data):
    return np.mean(data,axis=0)
# This function computes the centroid (mean feature vector) of a class

def class_spread(data):
    return np.std(data,axis=0)
# This function computes the intraclass spread using standard deviation

def interclass_distance(c1,c2):
    return np.linalg.norm(c1-c2)
# This function computes the Euclidean distance between two class centroids

def main():
    # Load iris dataset (all as strings first)
    data=np.genfromtxt(
        r"C:\EVERYTHING\AIE\2nd year\4th Sem\ML\Assignments\Lab2\iris.csv",
        delimiter=",",
        skip_header=1,
        dtype=str
    )

    # Split features and labels
    features=data[:,:4].astype(float)
    labels=data[:,4]

    # Take two classes
    class1=features[labels=="Iris-setosa"]
    class2=features[labels=="Iris-versicolor"]

    centroid1=class_centroid(class1)
    centroid2=class_centroid(class2)

    spread1=class_spread(class1)
    spread2=class_spread(class2)

    distance=interclass_distance(centroid1,centroid2)

    print("Class1 Mean:",centroid1)
    print("Class1 Spread:",spread1)
    print("Class2 Mean:",centroid2)
    print("Class2 Spread:",spread2)
    print("Interclass Distance:",distance)

if __name__=="__main__":
    main()
```

```python
# A3. Take any feature from your dataset. Observe the density pattern for that feature by plotting the histogram. Use buckets (data in ranges) for histogram generation and study. Calculate the mean and variance from the available data. (Suggestion: numpy.histogram()gives the histogram data. Plot of histogram may be achieved with matplotlib.pyplot.hist())
import numpy as np
import matplotlib.pyplot as plt

def mean_value(data):
    return np.sum(data)/data.shape[0]
# This function computes the mean of a one-dimensional data array

def variance_value(data):
    mu=mean_value(data)
    return np.sum((data-mu)**2)/data.shape[0]
# This function computes the variance of a one-dimensional data array

def main():
    # Load iris dataset
    data=np.genfromtxt(
        r"C:\EVERYTHING\AIE\2nd year\4th Sem\ML\Assignments\Lab2\iris.csv",
        delimiter=",",
        skip_header=1,
        dtype=str
    )

    # Extract sepal length feature
    sepal_length=data[:,0].astype(float)

    # Histogram data using buckets
    hist,bins=np.histogram(sepal_length,bins=10)

    print("Histogram Values:",hist)
    print("Bin Ranges:",bins)

    # Mean and variance
    mean=mean_value(sepal_length)
    variance=variance_value(sepal_length)

    print("Mean:",mean)
    print("Variance:",variance)

    # Plot histogram
    plt.hist(sepal_length,bins=10)
    plt.xlabel("Sepal Length")
    plt.ylabel("Frequency")
    plt.title("Histogram of Sepal Length")
    plt.show()

if __name__=="__main__":
    main()
```

```python
# A4. Take any two feature vectors from your dataset. Calculate the Minkwoski distance with p from 1 to 10. Make a plot of the distance and observe the nature of this graph. Develop your own function for calculation of the Minkowski distance metric.
import numpy as np
import matplotlib.pyplot as plt

def minkowski_distance(x,y,p):
    return np.sum(np.abs(x-y)**p)**(1/p)
# This function computes the Minkowski distance between two feature vectors for a given p value

def main():
    # Load iris dataset
    data=np.genfromtxt(
        r"C:\EVERYTHING\AIE\2nd year\4th Sem\ML\Assignments\Lab2\iris.csv",
        delimiter=",",
        skip_header=1,
        dtype=str
    )

    features=data[:,:4].astype(float)

    # Take any two feature vectors
    x=features[0]
    y=features[1]

    p_values=np.arange(1,11)
    distances=[]

    for p in p_values:
        distances.append(minkowski_distance(x,y,p))

    # Plot distance vs p
    plt.plot(p_values,distances,marker='o')
    plt.xlabel("p value")
    plt.ylabel("Minkowski Distance")
    plt.title("Minkowski Distance vs p")
    plt.grid(True)
    plt.show()

    # Print values
    for i in range(len(p_values)):
        print("p =",p_values[i],"Distance =",distances[i])

if __name__=="__main__":
    main()
```
